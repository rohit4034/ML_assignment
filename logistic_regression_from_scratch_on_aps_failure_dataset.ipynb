```python
def binary_cross_entropy(theta0, theta, X, y, lam):
    # Add regularization term to the loss function
    reg_term = (lam / (2 * X.shape[0])) * np.sum(np.square(theta[1:]))
    # Compute the binary cross entropy loss
    z = np.dot(X, theta) + theta0
    y_pred = 1 / (1 + np.exp(-z))
    loss = (-y * np.log(y_pred) - (1 - y) * np.log(1 - y_pred)).mean() + reg_term
    return loss
```

```python
def del_L_by_del_theta0(theta0, theta, X, y):
    z = np.dot(X, theta) + theta0
    y_pred = 1 / (1 + np.exp(-z))
    grad = (y_pred - y).mean()
    return grad
```

```python
def del_L_by_del_theta(theta0, theta, X, y, lam):
    # Add regularization term to the gradient
    reg_term = (lam / X.shape[0]) * theta
    z = np.dot(X, theta) + theta0
    y_pred = 1 / (1 + np.exp(-z))
    grad = np.dot(X.T, y_pred - y) / X.shape[0] + reg_term
    return grad
```

```python
def binary_logistic_regression(X, y, theta0_initial, theta_initial, lam=0.001, batch_size=32, max_iter=1500, epsilon=0.0001, tol=0.000001):
    counter = 0
    while counter <= max_iter:
        # Select a batch of samples
        indices = np.random.choice(X.shape[0], batch_size, replace=False)
        X_batch = X[indices]
        y_batch = y[indices]
        # Compute gradients and update weights
        theta0_final = theta0_initial - (epsilon * del_L_by_del_theta0(theta0_initial, theta_initial, X_batch, y_batch))
        theta_final = theta_initial - (epsilon * del_L_by_del_theta(theta0_initial, theta_initial, X_batch, y_batch, lam))
        # Check for convergence
        if np.abs(binary_cross_entropy(theta0_final, theta_final, X, y, lam) - binary_cross_entropy(theta0_initial, theta_initial, X, y, lam)) < tol:
            break
        print("The value of binary Cross Entropy at Epoch # {} is {} ".format(counter, binary_cross_entropy(theta0_initial, theta_initial, X, y, lam)))
        theta0_initial = theta0_final
        theta_initial = theta_final
        counter += 1
```

```python
binary_logistic_regression(X,y,theta0_initial,theta_initial,lam=0.001,batch_size=32, max_iter= 1500, epsilon=0.00001, tol = 0.000001)
```
