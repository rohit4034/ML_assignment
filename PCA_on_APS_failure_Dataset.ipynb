```python
# Find index of features with high correlation
high_corr = set()
for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if corr_matrix.iloc[i, j] >= 0.7:
            colname = corr_matrix.columns[i]
            high_corr.add(colname)

# the columns will not be reduced correctly with this technique, so we now apply pca on it
```

```python
# Calculate the mean and standard deviation of the data
mean = np.mean(X, axis=0)
std = np.std(X, axis=0)

# Standardize the data
X_scaled = (X - mean) / std
```

```python
# Calculate the covariance matrix of the standardized data
# cov_matrix is a m x m matrix, where m is the number of features in the dataset.
cov_matrix = np.cov(X_scaled.T)
```

```python
cov_matrix.shape
```

```python
null_values = np.isnan(cov_matrix)
```

```python
mean_value = np.nanmean(cov_matrix)

# replace the NaN values with the mean value
cov_matrix[np.isnan(cov_matrix)] = mean_value
```

```python
np.isnan(cov_matrix).sum().sum()
```

```python
# Calculate the eigenvectors and eigenvalues of the covariance matrix
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

```

```python
print(eigenvalues.shape)
print(eigenvectors.shape)
```

```python
print(eigenvalues)
print(eigenvectors)
```

```python
var=[]
for i in range(len(eigenvalues)):
  var.append(eigenvalues[i]/np.sum(eigenvalues))
var.sort(reverse=True)
print(np.sum(var),var,"\n")
print(np.sum(var[0:55]))
# now we can conclude that, 95 percent of information is coming from 55 columns, now we can use these 55 columns in our furthure anal
```

```python
  # Sort the eigenvalues and eigenvectors in descending order of eigenvalues
indices = np.argsort(eigenvalues)[::-1]
eigenvectors = eigenvectors[:, indices]
eigenvalues = eigenvalues[indices]
```

```python
# Select the first num_components eigenvectors
#  as we have seenm, 55 is our num component
eigenvectors = eigenvectors[:, :55]
```
